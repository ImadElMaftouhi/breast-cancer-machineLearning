#### e. ROC curve 

La courbe ROC (Receiver Operating Characteristic) illustre les performances d'un modÃ¨le de classification binaire en fonction de diffÃ©rents seuils de dÃ©cision. Elle reprÃ©sente le taux de vrais positifs (sensibilitÃ©) par rapport au taux de faux positifs (1 - spÃ©cificitÃ©) lorsque le seuil de classification des instances positives varie.

Les principaux aspects de la courbe ROC sont les suivants

1) La courbe illustre l'Ã©quilibre entre l'identification correcte des instances positives (sensibilitÃ©) et le rejet correct des instances nÃ©gatives (spÃ©cificitÃ©). Toute amÃ©lioration de la sensibilitÃ© s'accompagne gÃ©nÃ©ralement d'une rÃ©duction de la spÃ©cificitÃ©.

2) **Ã‰valuation de la prÃ©cision:** Un modÃ¨le performant prÃ©sente une courbe qui suit de prÃ¨s les limites gauche et supÃ©rieure de l'espace ROC. Cela indique une plus grande prÃ©cision globale pour diffÃ©rents niveaux de seuil.

3) **Comparaison avec la ligne diagonale:** La proximitÃ© de la courbe avec la ligne diagonale Ã  45 degrÃ©s, qui reprÃ©sente le hasard, sert d'indicateur de la capacitÃ© de discrimination du modÃ¨le. Une plus grande proximitÃ© avec cette ligne indique une prÃ©cision rÃ©duite, la performance du modÃ¨le se rapprochant de celle d'une supposition alÃ©atoire.

4) **Aire sous la courbe (AUC):** L'AUC quantifie la prÃ©cision globale du modÃ¨le. Une valeur de SSC plus Ã©levÃ©e signifie une meilleure discrimination entre les instances positives et nÃ©gatives en fonction de diffÃ©rents seuils, un classificateur parfait produisant une SSC de 1.

#### f. Cross validation ~ validation croisÃ©e

Avant de nous plonger dans les dÃ©tails de l'optimisation des hyperparamÃ¨tres, comprenons d'abord le concept de validation croisÃ©e.

La performance du modÃ¨le entraÃ®nÃ© dÃ©pend de la maniÃ¨re dont les donnÃ©es sont divisÃ©es. Elle peut ne pas Ãªtre reprÃ©sentative de la capacitÃ© du modÃ¨le Ã  gÃ©nÃ©raliser.

La solution est la validation croisÃ©e.

La validation croisÃ©e est une technique permettant d'Ã©valuer les modÃ¨les prÃ©dictifs en divisant l'Ã©chantillon d'origine en un ensemble d'entraÃ®nement pour former le modÃ¨le, et un ensemble de test pour l'Ã©valuer.

Dans la validation croisÃ©e en k-fold, l'Ã©chantillon d'origine est divisÃ© de maniÃ¨re alÃ©atoire en k sous-Ã©chantillons de taille Ã©gale. Parmi les k sous-Ã©chantillons, un seul sous-Ã©chantillon est conservÃ© comme donnÃ©es de validation pour tester le modÃ¨le, et les k-1 sous-Ã©chantillons restants sont utilisÃ©s comme donnÃ©es d'entraÃ®nement. Le processus de validation croisÃ©e est ensuite rÃ©pÃ©tÃ© k fois (les plis), chaque sous-Ã©chantillon Ã©tant utilisÃ© exactement une fois comme donnÃ©es de validation. Les k rÃ©sultats des plis peuvent ensuite Ãªtre moyennÃ©s (ou combinÃ©s d'une autre maniÃ¨re) pour produire une estimation unique. L'avantage de cette mÃ©thode est que toutes les observations sont utilisÃ©es Ã  la fois pour l'entraÃ®nement et la validation, et chaque observation est utilisÃ©e exactement une fois pour la validation.

#### g. Optimisation

La valeur de k (c'est-Ã -dire 2) que nous avons sÃ©lectionnÃ©e ci-dessus a Ã©tÃ© choisie en observant la courbe de prÃ©cision par rapport au nombre de voisins. C'est une mÃ©thode primitive d'optimisation des hyperparamÃ¨tres.

Il existe une meilleure faÃ§on de le faire qui implique :

1) Essayer diffÃ©rentes valeurs d'hyperparamÃ¨tres

2) Les ajuster tous sÃ©parÃ©ment

3) VÃ©rifier la performance de chacun

4) Choisir celui qui performe le mieux

5) Utiliser la validation croisÃ©e Ã  chaque fois

Scikit-learn fournit une mÃ©thode simple pour y parvenir en utilisant GridSearchCV, c'est-Ã -dire la recherche de grille avec validation croisÃ©e.

```python
param_grid = {'n_neighbors':np.arange(1,50)}
```

```python
knn = KNeighborsClassifier()

X_np = X.values

Y_np = Y.to_numpy()

knn_cv = GridSearchCV(knn, param_grid, cv=10)
knn_cv.fit(X_np,Y_np)
```

Ce code utilise la classe `KNeighborsClassifier` de Scikit-learn pour crÃ©er un classificateur KNN (k plus proches voisins). Ensuite, il utilise `GridSearchCV` pour effectuer une recherche sur une grille d'hyperparamÃ¨tres afin de trouver les meilleurs paramÃ¨tres pour le classificateur KNN. 

Voici une explication Ã©tape par Ã©tape :

1) `KNeighborsClassifier()`: CrÃ©e une instance du classificateur KNN.

2) `param_grid`: C'est un dictionnaire ou une liste de dictionnaires contenant les valeurs Ã  tester pour chaque hyperparamÃ¨tre du classificateur KNN.

3) `cv=10`: Indique que la validation croisÃ©e Ã  10 plis sera utilisÃ©e pour Ã©valuer la performance de chaque combinaison d'hyperparamÃ¨tres.

4) `knn_cv.fit(X,Y)`: Applique la recherche de grille avec validation croisÃ©e sur les donnÃ©es d'entraÃ®nement `X` et les Ã©tiquettes cibles `Y`. Cette mÃ©thode ajuste le modÃ¨le KNN Ã  chaque combinaison d'hyperparamÃ¨tres et Ã©value sa performance Ã  l'aide de la validation croisÃ©e.

Une fois que `fit()` est terminÃ©, `knn_cv` contient le modÃ¨le ajustÃ© avec les meilleurs hyperparamÃ¨tres trouvÃ©s par la recherche sur la grille.

```python
knn_cv.best_score_
```

### 5.1.2 Arbre de decision

#### a. Hyperparametres

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
max_depths = np.arange(1, 10)
criteria = ['gini', 'entropy']

plt.figure(figsize=(15, 5))

for criterion in criteria:
    train_accuracy = np.empty(len(max_depths))
    test_accuracy = np.empty(len(max_depths))
    
    for i, depth in enumerate(max_depths):
        dtc = DecisionTreeClassifier(criterion=criterion, max_depth=depth, random_state=1)
        dtc.fit(X_train, Y_train)
        train_accuracy[i] = dtc.score(X_train, Y_train)
        test_accuracy[i] = dtc.score(X_test, Y_test)
    
    sns.lineplot(x=max_depths, y=test_accuracy, label=f'Test Accuracy ({criterion})')
    sns.lineplot(x=max_depths, y=train_accuracy, label=f'Train Accuracy ({criterion})')

plt.xlabel('Profondeur maximale')
plt.ylabel('PrÃ©cision')
plt.title('PrÃ©cision de la formation et du test pour DecisionTreeClassifier avec diffÃ©rents critÃ¨res et profondeurs maximales')
plt.legend()
plt.show()
```

        On choisit max_depth = 7

#### b. Elaboration de module

```python
DTC = DecisionTreeClassifier(criterion= 'gini', max_depth=7, random_state=1)
DTC.fit(X_train, Y_train)
```

```python
print('La precision du modele DTC avec max_depth = 7 est : ', DTC.score(X_test, Y_test)) 
```

Contrairement Ã  DTC.score, qui est une mÃ©thode de l'objet modÃ¨le, accuracy_score est une fonction autonome qui prend deux arguments : les Ã©tiquettes vraies (vÃ©ritÃ© terrain) et les Ã©tiquettes prÃ©dites.
Bien que DTC.score et accuracy_score soient tous deux utilisÃ©s pour Ã©valuer la prÃ©cision d'un classificateur, DTC.score est gÃ©nÃ©ralement utilisÃ© pour obtenir la prÃ©cision directement Ã  partir de l'objet modÃ¨le, tandis que accuracy_score est utilisÃ© pour calculer la prÃ©cision sur la base des Ã©tiquettes vraies et prÃ©dites sÃ©parÃ©ment. Ils servent essentiellement le mÃªme objectif, mais dans des contextes lÃ©gÃ¨rement diffÃ©rents.

```python
DTC_pred = DTC.predict(X_test)
print("Precision du module est : ", accuracy_score(Y_test, DTC_pred))
```

#### g. Optimisation 

```python
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': np.arange(1,50),
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

clf = DecisionTreeClassifier(random_state=1)

grid_search = GridSearchCV(clf, param_grid, cv=5, scoring='accuracy')

grid_search.fit(X_train, Y_train)

best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Score:", best_score)

# Get the best model
best_model = grid_search.best_estimator_

# Evaluate the best model
test_accuracy = best_model.score(X_test, Y_test)
print("Test Accuracy of the Best Model:", test_accuracy)
```

#### g. Finalisation de modele

```python
DTC = DecisionTreeClassifier(criterion='gini', max_depth= 6 , min_samples_leaf=1, min_samples_split=2)

DTC.fit(X_train, Y_train)

# Predictions
DTC_pred = DTC.predict(X_test)

print('-'*15, 'Les metriques du modele DTC','-'*15)

print(f'Le score du modele est : {DTC.score(X_test, Y_test)}')
print(f'Accuracy: {accuracy_score(Y_test, DTC_pred)}')
print(f'Precision: {precision_score(Y_test, DTC_pred)}')
print(f'Recall: {recall_score(Y_test, DTC_pred)}')
print(f'F1 Score: {f1_score(Y_test, DTC_pred)}')
print('Confusion Matrix:')
print(confusion_matrix(Y_test, DTC_pred))
print('-'*50)

print(classification_report(Y_test, DTC_pred))


# Calculating necessary values for the ROC curve
Y_pred_prob = DTC.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(Y_test, Y_pred_prob)


plt.figure(figsize=(15, 5))

plt.plot([0, 1], [0, 1], 'k--')

sns.lineplot(x=fpr, y=tpr)

plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve pour le modele KNN apres optimisation')

plt.show()
```

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
C_values = np.arange(1, 100)

plt.figure(figsize=(15, 20))

for index, kernel in enumerate(kernels):
    train_accuracy = np.empty(len(C_values))
    test_accuracy = np.empty(len(C_values))
    
    for i, C in enumerate(C_values):
        svc = SVC(kernel=kernel, C=C)
        svc.fit(X_train, Y_train)
        svc_pred = svc.predict(X_test)
        train_accuracy[i] = svc.score(X_train, Y_train)
        test_accuracy[i] = svc.score(X_test, Y_test)
    
    plt.subplot(4,1,index+1)        
    sns.lineplot(x=C_values, y=test_accuracy, label=f'Test Accuracy ({kernel})')
    sns.lineplot(x=C_values, y=train_accuracy, label=f'Train Accuracy ({kernel})')
    plt.title(f'PrÃ©cision de la formation et du test pour SVC ( kernel = {kernel} ) ')
    plt.xlabel('Valeurs de C')
    plt.ylabel('PrÃ©cision')
    
plt.legend()
plt.show()
```

    On remarque que notre model SVC est plus performantes avec le kernel linear par default, on peut juste proposer C = 60

#### b. Elaboraton de modele

```python
gamma_values = ['low', 'high']
gamma_results = {}

for g in gamma_values:
    if g == 'low':
        gamma_val = 0.001  # Choose a low gamma value
    else:
        gamma_val = 1.0    # Choose a high gamma value
        
    svc = SVC(kernel='linear', C=60, gamma=gamma_val)  # SVC with linear kernel
    svc.fit(X_train, Y_train)
    svc_pred = svc.predict(X_test)
    
    gamma_results[g] = {
        'score': svc.score(X_test, Y_test), 
        'precision': precision_score(Y_test, svc_pred), 
        'recall': recall_score(Y_test, svc_pred), 
        'f1_score': f1_score(Y_test, svc_pred)
    }

# Create a bar plot to visualize the results
print(gamma_results)
plt.figure(figsize=(10, 6))
sns.barplot(data=pd.DataFrame(gamma_results).T.reset_index(), x='index', y='score', hue='index')
plt.title('Performance du modÃ¨le SVC (linÃ©aire, C=60) en fonction de gamma')
plt.xlabel('Gamma')
plt.ylabel('Score')
plt.show()
```

    Dans notre example, la valeur de gamma n'influence pas notre performance car le performance est deja parfait
